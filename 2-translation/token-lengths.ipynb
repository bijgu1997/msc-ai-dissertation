```python
from transformers import AutoTokenizer
```

```python
tokenizer = AutoTokenizer.from_pretrained("MLRS/BERTu")
```

```python
def tokenize(tokenizer, sentences):
  return [tokenizer.tokenize(x) for x in sentences]
```

```python
snli_premise_tokenized = tokenize(tokenizer, snli_premise)
snli_hypothesis_tokenized = tokenize(tokenizer, snli_hypothesis)
mnli_premise_tokenized = tokenize(tokenizer, mnli_premise)
mnli_hypothesis_tokenized = tokenize(tokenizer, mnli_hypothesis)
```

```python
len_snli_premise_tokenized = [len(x) for x in snli_premise_tokenized]
len_snli_hypothesis_tokenized = [len(x) for x in snli_hypothesis_tokenized]
len_mnli_premise_tokenized = [len(x) for x in mnli_premise_tokenized]
len_mnli_hypothesis_tokenized = [len(x) for x in mnli_hypothesis_tokenized]
```

```python
len_snli_tokenized = [sum(x)+3 for x in zip(len_snli_premise_tokenized, len_snli_hypothesis_tokenized)]
len_mnli_tokenized = [sum(x)+3 for x in zip(len_mnli_premise_tokenized, len_mnli_hypothesis_tokenized)]
```

```python
def describe_tokens(name, input_list):
  print(f"Dataset [{name}]")
  print(f"Max Token Length [{max(input_list)}]")
  print(f"Min Token Length [{min(input_list)}]")
  print(f"Average Token Length [{sum(input_list)//len(input_list)}]")
```

```python
describe_tokens("SNLI",len_snli_tokenized)
```

```python
describe_tokens("MNLI",len_mnli_tokenized)
```
